### (3)for pass/fail column

```python
# checking how many students are failed in exam
# value count
data_copy["Pass/Fail"].value_counts()
```

```python
# plotting the graph to get the more clear view for it.
plt.figure(figsize=(6,4))
sb.histplot(data_copy["Pass/Fail"])
```

* We observe that more than 300 students failed exam and less than 200 students  passed the exam.

```python
# plotting the relational graph
sb.relplot(y="Previous Exam Score",x="Study Hours",data=data_copy)
```

*The relation between Study Hours and their score is normal.

```python
sb.catplot(x="Pass/Fail",y="Study Hours",data=data_copy,kind="box")
```

* The student whose study hours is less than 6 hours had failed in exam.
* The student whose study hours is greater than 6 hours had passed in exam.

```python
sb.relplot(y="Previous Exam Score",x="Study Hours",hue="Pass/Fail",data=data_copy)
```

* we observe that the sudents whose study hours is greater than 6 and  the score  is greater than 60 had passed in exam.
